## Real time data for sales in a time span. 
## Working on a model to forecast the sales in coming year
## LSTM and RNN concepts are picked to complete the task

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('RSCCASN.csv')
```

```python
df.info()
```

```python
# data parse can be added while reading itself

df = pd.read_csv('RSCCASN.csv', parse_dates= True, index_col='DATE' )
```

```python
df.info()
```

```python
df.columns=['sales']
```

```python
df.head()
```

```python
df.plot(figsize=(10,8))
```

```python
df.shape
```

```python
# now need to split the data and need to select the test size to consider.
# as this is yearly data, to be able to cover all information, will take test size as 18

test_size = 18
test_ind = len(df)- test_size

# will split the dataset from test_ind in train and test

test_df = df.iloc[test_ind:]
train_df = df.iloc[:test_ind]
```

```python
test_df.shape
```

```python
train_df.shape
```

#### Scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python

scaler.fit(train_df)
```

```python
scaled_train = scaler.transform(train_df)
scaled_test = scaler.transform(test_df)
```

#### Timeseries Generator

```python
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
```

```python
len(test_df)
```

```python
# batch size should be less than the actual size of test dataset

len_output = 12
gen = TimeseriesGenerator(scaled_train, scaled_train, length= len_output, batch_size=1)
```

```python
X, y = gen[0]
```

```python
X, y
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense, LSTM
```

```python
n_features = 1
```

```python
model = Sequential()
model.add(LSTM(100,activation = 'relu', input_shape=(len_output,n_features)))
model.add(Dense(1))
model.compile(optimizer='adam',loss='mse')
```

```python
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stops = EarlyStopping(monitor = 'val_loss', patience = 2)
```

```python
val_gen = TimeseriesGenerator( scaled_test, scaled_test, length= len_output, batch_size=1)
```

```python
model.fit_generator(gen, epochs=20, validation_data=val_gen,callbacks=[early_stops])
```

```python
pd.DataFrame(model.history.history).plot()
```

```python
# evaluate the model for complete test data

test_pred = []

eval_batch = scaled_train[-len_output:]

eval_batch=eval_batch.reshape((1,len_output,n_features))

for i in range(len(test_df)):
    
    pred = model.predict(eval_batch)[0]
    
    test_pred.append(pred)
    
    eval_batch = np.append(eval_batch[:,1:,:],[[pred]],axis=1)
```

```python
test_pred = scaler.inverse_transform(test_pred)
```

```python
test_df['prediction']= test_pred
```

```python
test_df.head()
```

```python
test_df.plot()
```

# Look for forecast

```python
scaler_full = MinMaxScaler()
scaled_full_data = scaler_full.fit_transform(df)
```

```python
len_output = 12
gen_full = TimeseriesGenerator(scaled_full_data,scaled_full_data,
                              length = len_output, batch_size=1)
```

```python
model = Sequential()
model.add(LSTM(100,activation = 'relu', input_shape=(len_output,n_features)))
model.add(Dense(1))
model.compile(optimizer='adam',loss='mse')

model.fit_generator(gen_full, epochs= 10 )
```

```python
# evaluate the model for complete test data

period = 12
forecasting = []

eval_batch = scaled_train[-len_output:]

eval_batch=eval_batch.reshape((1,len_output,n_features))

for i in range(period):
    
    pred = model.predict(eval_batch)[0]
    
    forecasting.append(pred)
    
    eval_batch = np.append(eval_batch[:,1:,:],[[pred]],axis=1)
```

```python
forecasting = scaler_full.inverse_transform(forecasting)
```
